```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('BostonHousing.csv')
```

```python
df.isnull().sum()
```

```python
X=df[['rm','lstat']].values
y=df['medv'].values
```

```python
def compute_loss(X,y,weight,bias):
    prediction = X @ weight + bias
    return np.mean((prediction - y)**2)
```

```python
def gradient_descent(X_batch,y_batch,weight,bias,learning_rate):
    m = len(X_batch)
    predictions = X_batch @ weight + bias
    error = predictions - y_batch

    gradient_weight = (2/m)* np.sum(X_batch.T @ error)
    gradient_bias = (2/m) * np.sum(error)

    weight -= learning_rate * gradient_weight
    bias -= learning_rate * gradient_bias

    return weight, bias
```

```python
def train_model(X,y,learning_rate=0.0001,epochs=100):
    np.random.seed(42)
    weight = np.random.randn(2) *0.001
    bias = np.random.rand() *0.001
    loss_history =[]
    for epoch in range(epochs):
        # indices  = len(y)
        indices = np.arange(len(y))  # Generate valid indices
        # indices = np.random.permutation(len(y))  # Shuffle indices properly

        X_full = X[indices]
        y_full = y[indices]
        for i in range(0,len(y)):
            print(i)
            X_batch = X_full[i:len(y)]
            y_batch = y_full[i:len(y)]
            weight, bias = gradient_descent(X_batch, y_batch,weight,bias, learning_rate)
            loss = compute_loss(X,y,weight,bias)
            loss_history.append(loss)
    return loss_history
```

```python
low=train_model(X,y,learning_rate=0.0001,epochs=100)
# print(low)
```

```python
plt.figure(figsize=(12,5))
plt.plot(low,label='lr 0.01',color='blue')
# plt.plot(medium,label='lr 0.001',color='green')
# plt.plot(high,label='lr 0.001',color='red')
plt.xlabel('epochs')
plt.ylabel('loss_history')
plt.title('loss curve')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def train_model(X,y,learning_rate=0.001, batch_size=20, epochs=500):
    np.random.seed(42)
    weight = np.random.randn(2) *0.001
    bias = np.random.rand() *0.001
    loss_history =[]
    for epoch in range(epochs):
        indices  = np.random.permutation(len(y))
        X_shuffled = X[indices]
        y_shuffled = y[indices]
        for i in range(0,len(y),batch_size ):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]
            weight, bias = gradient_descent(X_batch, y_batch,weight,bias, learning_rate)
        loss = compute_loss(X,y,weight,bias)
        loss_history.append(loss)
    return loss_history
```
