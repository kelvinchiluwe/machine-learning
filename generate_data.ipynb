```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
np.random.seed(42)
X=np.random.rand(100,1)*1000
y=2*X+3+np.random.randn(100,1)*50
data=pd.DataFrame({'house_size':X.flatten(),'price':y.flatten()})
```

```python
data.describe()
```

```python
X=data[['house_size']]
y=data['price']
```

```python
plt.scatter(X,y,color='green')
# plt.plot(X,y,color='yellow')
plt.xlabel('house_size')
plt.ylabel('price')
plt.title('relationship between price and house_size')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,X_test=train_test_split(X,y,train_size=0.8,random_state=11)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
y_predict=model.predict(X_test)
print(y_predict)
```

```python
r2=r2_score(y_test,y_predict)
print(r2)
```

```python
 meanAbsolute_Error= mean_absolute_error(y_test,y_predict)
print(meanAbsolute_Error)
```

```python
Mean_Squared_error=mean_squared_error(y_test,y_predict)
print(Mean_Squared_error)
```

```python
bias=model.intercept_
print(f' the bias is {bias}')
weight=model.coef_
print(f' the weight is {weight}')
```
