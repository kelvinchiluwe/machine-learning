```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
temperature=np.array([60,65,70,75,80,85,90,95,100])
rain=np.array([0,0,0,0,1,1,1,1,1])
temperature
rain
```

```python
# X=(X-X.mean())/X.std()
weight=np.random.randn()
bias=np.random.randn()
np.random.seed(45)
```

```python
def log_odds(weight,bias,X,y):
    return weight*X + bias
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
    
```

```python
def gradient_descent(X,y,weight,bias,learning_rate=0.00001):
    m=len(X)
    z=log_odds(weight,bias,X,y)
    prediction=sigmoid(z)
    error=prediction - y


    gradient_weight=(2/m)*np.sum(error*X)
    gradient_bias=(2/m)*np.sum(error)


    weight -= learning_rate*gradient_weight
    bias -=learning_rate*gradient_bias
    return weight,bias
```

```python
# def training_log_sigmoid(X,y,weight,bias,learning_rate):
epochs=100
losse=[]
for epoch in range(epochs):
    weight,bias= gradient_descent(temperature,rain,weight,bias,0.00001)
    z=log_odds(weight,bias,temperature,rain)
    prediction=sigmoid(z)
    loss= -np.mean(rain*np.log(prediction) + (1 - rain) * np.log(1 - prediction))
    losse.append(loss)
    print(losse)
```

```python
# training_log_sigmoid(temperature,rain,weight,bias,0.001)
plt.plot(losse,color='red')
plt.xlabel('epochs')
plt.ylabel('rain or no rain')
plt.title('loss_curve_logistics')
plt.grid(True)
plt.show()
```
